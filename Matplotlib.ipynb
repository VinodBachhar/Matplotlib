```python
# start angle
x=np.array([25,40,15,20])
y=["apple","mango","jackfruit","watermelon"]
plt.pie(x,labels=y)
plt.show()
```

```python
# Explode
x=np.array([25,40,15,20])
y=["apple","mango","jackfruit","watermelon"]
z=[0.2,0,0,0]
plt.pie(x, labels=y,explode=z)
plt.show()
```

```python
# Shadow
x=np.array([25,40,15,20])
y=["apple","mango","jackfruit","watermelon"]
z=[0.2,0,0,0]
plt.pie(x, labels=y,explode=z,shadow=True)
plt.show()
```

```python
names=np.array(["akash","asif","riyad","hasan","farid","salman"])
marks=np.array([545,754,608,733,504,722])
plt.figure(figsize=(12,7))
plt.title('Total mark obtained by the students',fontsize=20,color='r')
plt.xlabel('Name of the students',fontsize=15,color='r')
plt.ylabel('Mark',fontsize=15,color='r')
plt.bar(names,marks)
plt.show()
```
